####  • Data after Handling missing data
New dataset is saved in customer_data_clean.csv file

```python
import pandas as pd
from pathlib import Path
from IPython.display import display, HTML

inp = Path(r"D:/Jouwana-BZU/5th Year/First Semester - 2025-2026/Machine Learning/Last_One_Assignment#1/customer_data.csv")
out = Path(r"D:/Jouwana-BZU/5th Year/First Semester - 2025-2026/Machine Learning/Last_One_Assignment#1/customer_data_clean.csv")

# Load input dataset
df = pd.read_csv(inp)

# Keep a copy of the original dataset
df_original = df.copy()

# Identify numeric and categorical columns
num_cols = df.select_dtypes(include=["number"]).columns
cat_cols = df.select_dtypes(exclude=["number"]).columns

# Handle missing values using median for numeric features
for c in num_cols:
    df[c] = df[c].fillna(df[c].median())

# Handle missing values using mode for categorical features
for c in cat_cols:
    if df[c].isna().any():
        m = df[c].mode(dropna=True)
        if len(m) > 0:
            df[c] = df[c].fillna(m.iloc[0])

# Save cleaned dataset to customer_data_clean.csv
df.to_csv(out, index=False)

# Subset for display (first 50 rows)
df_display = df.head(50)
df_original_display = df_original.head(50)

# Check missing values
display(HTML("<h3><b>Check Missing Values:</b></h3>"))
print(df.isnull().sum())

# Highlight cells that were originally NaN
def highlight_filled(s):
    mask = df_original_display[s.name].isna()
    return ['background-color: yellow' if v else '' for v in mask]

display(HTML("<h3><b>Handled Missing Values (highlighted in yellow):</b></h3>"))

# Apply highlighting and format numeric columns to 1 decimal
df_display.style.apply(highlight_filled).format(
    {col: "{:.1f}".format for col in num_cols}  # 1 decimal for numeric columns
)
```

## Justification:

- Median is robust to outliers for numeric features.

- Mode is appropriate for categorical features.

- Deleting rows might reduce data unnecessarily, so imputation is preferred.

### 3.2. Discussing strategies to handle outliers.

### 3.2.1. IQR Strategy with Median Replacement
To prevent these extreme points from skewing the analysis, the Interquartile Range (IQR) method was applied to determine acceptable lower and upper bounds. Any values falling outside these bounds were replaced with the median of the corresponding feature. Median replacement was chosen because it is robust to extreme values and helps maintain the central tendency of the data without removing potentially useful records. After handling outliers, updated box plots showed a more reasonable and consistent data distribution across numerical features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# Paths
in_path  = r"D:\Jouwana-BZU\5th Year\First Semester - 2025-2026\Machine Learning\Last_One_Assignment#1\customer_data_clean.csv"
out_path = r"D:\Jouwana-BZU\5th Year\First Semester - 2025-2026\Machine Learning\Last_One_Assignment#1\customer_data_handled_outliers.csv"

# Load dataset
df = pd.read_csv(in_path)

# Numeric columns only
num_cols = df.select_dtypes(include="number").columns

# Select numeric columns with more than 2 unique values (non-binary)
non_binary_cols = [col for col in num_cols if df[col].nunique(dropna=True) > 2]

# Handle outliers using median replacement
for col in non_binary_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median_val = df[col].median()
    
    # Replace outliers with median
    df.loc[df[col] < lower_bound, col] = median_val
    df.loc[df[col] > upper_bound, col] = median_val

# Save the handled outliers dataset
df.to_csv(out_path, index=False)

display(Markdown(f"<h2 style='text-align:center; color:black;'>Handled Outliers Using Median<br></h2>"))

# Numeric, non-binary columns only
num_cols = df.select_dtypes(include="number").columns
non_binary_cols = [c for c in num_cols if df[c].nunique(dropna=True) > 2]

# Create one large figure for all boxplots
plt.figure(figsize=(12, 6))  # Adjust width/height as needed

# Plot each feature in a separate subplot
for i, col in enumerate(non_binary_cols, 1):
    plt.subplot(1, len(non_binary_cols), i)
    df.boxplot(column=col)
    plt.title(col)
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

# 4. Feature Scaling:
• Normalizing numerical features using Z-score

Z-score normalization (standardization) was used to scale numerical features because it transforms them to have a mean of 0 and a standard deviation of 1, making features comparable regardless of their original scales.

```python
# Step 0: Import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Step 1: Load your preprocessed dataset (after missing data handling & outlier capping)
df = pd.read_csv("customer_data_handled_outliers.csv")  # replace with your path

# Step 2: Inspect numeric columns before scaling
numeric_cols = ['Age', 'Income', 'Tenure', 'SupportCalls']
print("First 5 rows before scaling:")
print(df[numeric_cols].head())

print("\nSummary statistics before scaling:")
print(df[numeric_cols].describe())

# Step 3: Standardize numeric features using Z-score
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Step 4: Inspect numeric columns after scaling
print("\nFirst 5 rows after scaling:")
print(df[numeric_cols].head())

# Step 5: Save the standardized dataset
output_path = "customer_data_standardized.csv" 
df.to_csv(output_path, index=False)

print(f"\nStandardized dataset saved to: {output_path}")

print("\nSummary statistics after scaling:")
print(df[numeric_cols].describe())

# Numeric, non-binary columns only
num_cols = df.select_dtypes(include="number").columns
non_binary_cols = [c for c in num_cols if df[c].nunique(dropna=True) > 2]

# Create one large figure for all boxplots
plt.figure(figsize=(12, 6))  # Adjust width/height as needed

# Plot each feature in a separate subplot
for i, col in enumerate(non_binary_cols, 1):
    plt.subplot(1, len(non_binary_cols), i)
    df.boxplot(column=col)
    plt.title(col)
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

### 5.1.2. Analyzing the distribution of categorical variables using bar plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv(r"D:\Jouwana-BZU\5th Year\First Semester - 2025-2026\Machine Learning\Last_One_Assignment#1\customer_data_handled_outliers.csv")

cat_cols = ['Gender', 'ProductType', 'ChurnStatus']

# Define colors for categories
color_map = {
    'Gender': ['skyblue', 'lightpink'],       # Male, Female
    'ProductType': ['lightgreen', 'orange'],  # Basic, Premium
    'ChurnStatus': ['lightcoral', 'lightblue']  # Churned, Stayed
}

# Define label descriptions for each category
label_map = {
    'Gender': ['Male', 'Female'],
    'ProductType': ['Basic', 'Premium'],
    'ChurnStatus': ['Churned', 'Stayed']
}

# Create a single figure with subplots
plt.figure(figsize=(18, 5))  # Large figure

for i, col in enumerate(cat_cols):
    plt.subplot(1, 3, i+1)  # 1x3 grid
    colors = color_map[col]
    labels = label_map[col]
    
    sns.countplot(x=df[col], hue=df[col], palette=colors, legend=False)
    plt.title(f'Bar Plot of {col}', fontsize=14, fontweight='bold')
    plt.xlabel(f"{col} Categories", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(ticks=range(len(labels)), labels=labels, fontsize=11)
    
    # Add color explanation on top of each subplot
    color_description = " → ".join([f"{labels[j]}: {colors[j]}" for j in range(len(labels))])
    plt.text(0.5, df[col].value_counts().max()*1.05, color_description, ha='center', fontsize=10, color='dimgray', transform=plt.gca().transData)

plt.tight_layout()
plt.show()
```

## 5.2. Bivariate Analysis:

### 5.2.1. Exploreing relationships between numerical features and the target variable (ChurnStatus) using box plots.
For numerical features vs. the target (ChurnStatus), the most suitable visualization is usually a box plot rather than a scatter plot, because ChurnStatus is binary (0/1). Box plots will clearly show how the distribution of each numeric feature differs for customers who churned vs. stayed.

```python
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# Load dataset with handled outliers
df = pd.read_csv(r"D:\Jouwana-BZU\5th Year\First Semester - 2025-2026\Machine Learning\Last_One_Assignment#1\customer_data_handled_outliers.csv")

# Numeric, non-binary columns only
num_cols = df.select_dtypes(include="number").columns
non_binary_cols = [c for c in num_cols if df[c].nunique(dropna=True) > 2]

target = 'ChurnStatus'

# Boxplots for each numeric feature grouped by ChurnStatus
for col in non_binary_cols:
    display(Markdown(f"**Boxplot of {col} grouped by {target}**"))
    
    plt.figure(figsize=(6,4))
    df.boxplot(column=col, by=target)
    plt.title(f"{col} vs {target}")
    plt.suptitle('')  # remove the default pandas title
    plt.xlabel(target)
    plt.ylabel(col)
    
    # Add custom x-axis tick labels: keep 0 and 1, but show meaning beside them
    plt.xticks(
        ticks=[1, 2], 
        labels=["0 (Stayed)", "1 (Churned)"], 
        fontsize=10
    )
    
    plt.tight_layout()
    plt.show()
```

### 5.2.2. Investigating relationships between categorical variables and the target using bar plots.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Categorical columns to analyze
cat_cols = ['Gender', 'ProductType']

# Create one large figure with subplots
plt.figure(figsize=(12, 5))  # adjust width/height as needed

for i, col in enumerate(cat_cols, 1):
    plt.subplot(1, len(cat_cols), i)  # 1 row, n columns
    
    # Compute proportions of ChurnStatus per category
    prop_df = df.groupby([col, 'ChurnStatus']).size().unstack(fill_value=0)
    prop_df = prop_df.div(prop_df.sum(axis=1), axis=0)  # normalize to proportions
    
    # Plot stacked bar
    prop_df.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], ax=plt.gca())
    
    plt.title(f"{col} vs ChurnStatus (Proportion)", fontsize=12)
    plt.xlabel(col)
    plt.ylabel("Proportion")
    plt.legend(title="ChurnStatus", labels=['Stayed (0)', 'Churned (1)'])

plt.tight_layout()
plt.show()
```

## 5.3. Correlation Analysis:

### 5.3.1. Create a correlation matrix to examine how numerical features relate to each other and to the target variable.
- A heatmap is the most common visualization for correlations.
- The correlation with ChurnStatus indicates which numeric features are most predictive:

    - Positive correlation → higher values of the feature are associated with churn.

    - Negative correlation → higher values of the feature are associated with staying.

    - Features with correlation close to 0 → weak linear relationship with churn.

    - Example discussion:

        - SupportCalls have a positive correlation with ChurnStatus → customers who call support more often are more likely to churn.

        - Tenure have a negative correlation → long-term customers are less likely to churn.

        - Income and Age  have weak correlations → they are less predictive individually.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"D:\Jouwana-BZU\5th Year\First Semester - 2025-2026\Machine Learning\Assignment#1\customer_data_handled_outliers.csv")

# Numeric columns including the target
num_cols = ['Age', 'Income', 'Tenure', 'SupportCalls', 'ChurnStatus']

# Compute correlation matrix
corr_matrix = df[num_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Matrix of Numeric Features", fontsize=16)
plt.show()
```

### 5.3.2. Discussing which features seem most predictive of churn.

```python
# Compute correlation with target
corr_with_target = df[num_cols].corr()['ChurnStatus'].sort_values(ascending=False)
print("Correlation of features with ChurnStatus:\n")
print(corr_with_target)
```

1. Features with highest positive correlation → more likely to be associated with churn:

    - Example: SupportCalls might have positive correlation → customers who make more support calls are more likely to churn.

2. Features with highest negative correlation → inversely associated with churn:

    - Example: Tenure might have negative correlation → long-term customers are less likely to churn.

3. Features with weak correlation (close to 0) → low predictive power individually:

    - Example: Age and Income might have small correlations → not strongly predictive on their own.

 - Conclusion

    - Most predictive numeric features for churn: SupportCalls (positive), Tenure (negative).

    - Less predictive numeric features: Age, Income.

    - For categorical features like Gender or ProductType, you can analyze their bar plots vs churn to check if they have a strong relationship with churn.
